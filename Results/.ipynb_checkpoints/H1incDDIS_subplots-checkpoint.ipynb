```python
import yaml
import numpy as np
import pandas as pd
import lhapdf as lh
import matplotlib
import matplotlib.pyplot as plt
#matplotlib.rcParams['text.usetex'] = True 
import MatplotlibSettings#COMPASS
from collections import deque
import copy
import os
from os import listdir
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# H1incDDIS

xpom = [0.001, 0.003, 0.01,0.03]
from pathlib import Path
import itertools
import re ##  regular expression package for searching in strings
FitFolder='./NNLO-inc-dijet/'
paths = [path for path in Path(FitFolder).resolve().glob("**/*PrintDataTheory_H1incDDIS_*.txt") if 'Dijet' not in path.stem]
#leaving out dijet data, keeping only DDIS
###############
f, axs = plt.subplots(2, 2, sharex = "row" , sharey='row' ,figsize=(14, 14))
plt.subplots_adjust(wspace = 0, hspace = 0.)
#############################################
pointcolor= 'g'
markerth=''
thlinestyle='-'
lw=1.5
alpha=1
bandcolor='#ffff33'
######
color,marker, markersize, markerfacecolor, markeredgecolor,markeredgewidth='k','o',6,'r','k',1
linewidth,capsize,capthick=1.5,5,1
for index ,path in enumerate(paths):
    fileas='f'+str(index)
    with open(str(path)) as fileas:
        print('parsing file:',path)
        preds = pd.read_csv(fileas, delim_whitespace=True)
        dirname = str(path.parents[0])
        print("hadi")
        dirname= dirname[len(str(path.parents[1]))+1:] # Returns immidiate parent folder name of file by removing the parents
        preds = preds.sort_values('beta',ascending=False)
        #######################################
        e1 = preds.loc[preds['xp'] == xpom[0]]
        e2 = preds.loc[preds['xp'] == xpom[1]]
        e3 = preds.loc[preds['xp'] == xpom[2]]
        e4 = preds.loc[preds['xp'] == xpom[3]]
        #######################################
        dflist=[e1,e2,e3,e4]
        for edf in enumerate(dflist):
            betaselects = edf[1]['beta'].unique()
            for element in enumerate(betaselects):
                factor = 3**(element[0])
                e = edf[1].loc[(edf[1]['beta'] == element[1])].copy()
            ##############################################
                xv    = e["Q2"].to_numpy()
                fv    = e["Theory"].to_numpy()* factor
                dfv   = e["ErrTot"].to_numpy()* factor
                datav = e["Data"].to_numpy()* factor
                ddata = e["ErrTot"].to_numpy()* factor
                ## HH sort by Q2
                permutation = xv.argsort()
                xv    = xv[permutation]
                fv    = fv[permutation]
                datav = datav[permutation]
                ddata = ddata[permutation]
            #########################################
                if   edf[0]==0:
                    j,i=0,0
                    axs[j,i].set_ylabel(r"\textbf{$3^i x_{I\!\!P}\sigma_{\mathrm{reduced}}$}")
                elif edf[0]==1:
                    j,i=0,1
                elif edf[0]==2:
                    j,i=1,0
                    #axs[j,i].set_xlabel(r"\textbf{$Q^2\ (\mathrm{GeV}^2)$}")
                else:
                    j,i=1,1
                    #axs[j,i].xaxis.set_label_coords(50, 1e-3)
                    axs[j,i].set_xlabel(r"\textbf{$Q^2\ (\mathrm{GeV}^2)$}")#, horizontalalignment='left', x=0.0)
                if j==0:
                    
                    plt.ylim(1e-2,1e2)
                    #plt.xlim(1e0,1e8)
                    axs[j,i].set_xlim([1e1, 2e2])

                    axs[j,i].xaxis.set_ticks_position('top')
                    axs[j,i].xaxis.set_ticks_position('both')
                else:
                    plt.ylim(1e-2,1e4)
                    plt.xlim(7.0,7e3)
            #########################################
                #print(f'i={i} , j={j}')
                axs[j,i].set_yscale('log')
                axs[j,i].set_xscale('log')
                center=axs[j,i].plot(xv, fv, color = pointcolor, ls = thlinestyle, lw = lw, marker=markerth)
                axs[j,i].annotate(r'$\beta='+str(element[1])+r'$',(np.amax(xv)*1.15,fv[np.where(xv == np.amax(xv))])\
                                 , fontsize=14)## HH: wrinting labels next to curves
                errband=axs[j,i].fill_between(xv, np.array(fv) + np.array(dfv), np.array(fv) - np.array(dfv),\
                                          color = bandcolor, alpha = alpha,edgecolor=pointcolor)

                dataplot=axs[j,i].errorbar(xv, datav, yerr=ddata, color = color,linewidth=linewidth\
                     ,marker=marker, markersize=markersize, markerfacecolor=markerfacecolor, markeredgecolor=markeredgecolor,markeredgewidth=1\
                     ,capsize=capsize,capthick=capthick,ls='none')
                if ((j==0) and (i==0)):
                    p2 = axs[j,i].fill(np.NaN, np.NaN, color=bandcolor, alpha=alpha,edgecolor=pointcolor)#### for combined legend only
                    plt.rcParams['legend.title_fontsize'] = 'xx-large'
                    axs[j,i].legend([(p2[0], center[0]),(dataplot,) ], [r'\textbf{'+"NNLO-dijet"+r'}',r'\textbf{'+str(re.search('_(.*)_cut',\
                        str(path.stem)).group(1))+r'}'],loc='best',title=r"\textbf{$x_{I\!\!P} = " + str(xpom[edf[0]])+r'$}' \
                        ,fontsize=20)
                else:
                    axs[j,i].legend(loc='best',title=r"\textbf{$x_{I\!\!P} = " + str(xpom[edf[0]])+r'$}' \
                        ,fontsize=20)
            #################################################################################################s
            ######################
                e = pd.DataFrame(None)
        edf = pd.DataFrame(None)

plt.savefig('./' + str(dirname) + str(re.search('_(.*)_cut',str(path.stem)).group(1)) + ".pdf")
plt.show()
```
