```python
from pathlib import Path
import itertools
import re ##  regular expression package for searching in strings
FitFolder='./'
paths = [path for path in Path(FitFolder).resolve().glob("**/*Print*.txt") if 'Dijet' not in path.stem]
#leaving out dijet data, keeping only DDIS
for index ,path in enumerate(paths):
    #print('opening folder:', path.parents[0])
    fileas='f'+str(index)
    #print(fileas)
    with open(str(path)) as fileas:
        print('parsing file:',path)
        preds = pd.read_csv(fileas, delim_whitespace=True)
        dirname = str(path.parents[0])
        dirname= dirname[len(str(path.parents[1]))+1:] # Returns immidiate parent folder name of file by removing the parents
        #print(path.stem)
        #if 'beta' in preds.columns:
        preds = preds.sort_values('beta')
        Q2selects = preds['Q2'].unique()
        xpselects = preds['xp'].unique()
        preds = preds[['Q2', 'xp', 'beta','Data','ErrTot','Theory']].copy()
        for element in itertools.product(Q2selects,xpselects):
            Q2select,xpselect=element[0],element[1]
            e = preds.loc[(preds['Q2'] == Q2select)&(preds['xp'] == xpselect)].copy()
            print(Q2select,xpselect,e.shape[0])
            if e.shape[0]<4:
                continue
            datav = e["Data"].to_numpy()
            ddata = e["ErrTot"].to_numpy()
            xv  = e["beta"].to_numpy()
            fv  = e["Theory"].to_numpy()
            dfv = e["ErrTot"].to_numpy()
            f, (ax1, ax2) = plt.subplots(2, 1, sharex = "all", gridspec_kw = dict(width_ratios = [1], height_ratios = [3, 1]),figsize=(9, 9))
            plt.subplots_adjust(wspace = 0, hspace = 0)
            #ax1.set_title(r"$Q = " + str(Q2select) + r"$ \textbf{GeV}, $x_{I\!P}=$"+r'$\textbf{'+str(xpselect)+r'}$')
            ax1.set_ylabel(r"\textbf{$x_{I\!P}\sigma_\mathrm{reduced}$}")
            ax1.set_xscale("log")
            ax2.set_xlabel(r"$\beta$")
            ax2.set_ylabel(r"\textbf{Theory/Data}", fontsize = 15)
            ax2.set_xscale("log")
            ax2.set_ylim([0.5, 1.4])
            pointcolor= 'g'
            markerth=''
            thlinestyle='-'
            lw=1.5
            center=ax1.plot(xv, fv, color = pointcolor, ls = thlinestyle, lw = lw, marker=markerth)
            color,marker, markersize, markerfacecolor, markeredgecolor,markeredgewidth='k','o',8,'r','k',1
            linewidth,capsize,capthick=1.5,5,1
            alpha=1
            dataplot=ax1.errorbar(xv, datav, yerr=ddata, color = color,linewidth=linewidth,marker=marker\
                                  , markersize=markersize, markerfacecolor=markerfacecolor, markeredgecolor=markeredgecolor,markeredgewidth=1\
                                  ,capsize=capsize,capthick=capthick,ls='none')
            bandcolor='#ffff33'
            errband=ax1.fill_between(xv, np.array(fv) + np.array(dfv), np.array(fv) - np.array(dfv), color = bandcolor, alpha = alpha,edgecolor=pointcolor)
            p2 = ax1.fill(np.NaN, np.NaN, color=bandcolor, alpha=alpha,edgecolor=pointcolor)#### for combined legend only
            ax2.plot(xv, np.array(fv)/np.array(datav), color = pointcolor, ls = thlinestyle, lw = lw, marker=markerth)
            ax2.errorbar(xv, datav/datav, yerr=ddata/datav, color = color,linewidth=linewidth,marker=marker\
                         , markersize=markersize, markerfacecolor=markerfacecolor, markeredgecolor=markeredgecolor,markeredgewidth=1\
                         ,capsize=capsize,capthick=capthick,ls='none')
            ax2.fill_between(xv, np.array(fv/datav) + np.array(dfv/datav), np.array(fv/datav) - np.array(dfv/datav)\
                             , color = bandcolor, alpha = alpha,edgecolor=pointcolor)
            plt.rcParams['legend.title_fontsize'] = 'xx-large'
            ax1.legend([(p2[0], center[0]),(dataplot,) ], [r'\textbf{'+str(dirname)+r'}',r'\textbf{'+str(re.search('_(.*)_cut',\
                    str(path.stem)).group(1))+r'}'],loc='best',title=r"$Q = " + str(Q2select) + r"$ \textbf{GeV}, $x_{I\!P}=$"+r'$\textbf{'+str(xpselect)+r'}$'\
                    ,fontsize=20)
            # Create plots folder
            if not os.path.exists(FitFolder + "/DDIS-Plot/"+str(dirname)):
                os.makedirs(FitFolder + "/DDIS-Plot/"+str(dirname))
            plt.savefig(FitFolder + "/DDIS-Plot/" +str(dirname)+'/'+ str(path.stem)+'_Q2_'+str(Q2select)+'_xP_'+str(xpselect) + ".pdf")
            plt.close()
        e = pd.DataFrame(None)
        pred = pd.DataFrame(None)
```
