
DiffDIS
=====

Diffractive Deep Inelastic Scattering

```python
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
from feynman import Diagram

fig = matplotlib.pyplot.figure(figsize=(10.,10.))
ax = fig.add_axes([0,0,1,1], frameon=False)

pom_style = dict(arrow=False, style = 'double',linewidth=7)
t_style = dict(style='elliptic',
               ellipse_excentricity=-1.2, ellipse_spread=.3,
               arrow=True,ellipse_position='down',lw=2)

diagram = Diagram(ax)
#diagram.text(.4,0.9,"Diffractive Deep Inelastic Scattering", fontsize=40)
in1  = diagram.vertex(xy=(.1,.85), marker='')
in2  = diagram.vertex(xy=(.1,.35), marker='')
out1 = diagram.vertex(xy=(.95,.95), marker='')
out2 = diagram.vertex(xy=(.95,.25), marker='')
v1   = diagram.vertex(xy=(.4,.375), marker='')
vqed = diagram.vertex(xy=(.4,.85), marker='')
v3   = diagram.vertex(xy=(.4,.55), marker='')
v4   = diagram.vertex(xy=(.4,0.5), marker='')
v5   = diagram.vertex(xy=(.425,0.525), marker=' ')
vx   = diagram.vertex(xy=(.55,.7), marker='')
v1p  = diagram.vertex(xy=(.425,.35),marker='')
v1pp = diagram.vertex(xy=(.375,.35),marker='')
v1t  = diagram.vertex(xy=(.3,.3),marker='')
v2t  = diagram.vertex(xy=(.52,.27),marker='')

l1plus = diagram.vertex(xy=(.95,.8), marker='')
l2plus = diagram.vertex(xy=(.95,.6), marker='')
l1minus = diagram.vertex(xy=(.95,.4), marker='')
l2minus = diagram.vertex(xy=(.95,.2), marker='')
outx1 = diagram.vertex(xy=(.7,.58), marker='')
outx2 = diagram.vertex(xy=(.7,.55), marker='')
outx3 = diagram.vertex(xy=(.7,.52), marker='')
outX  = diagram.vertex(xy=(.8,.7), marker='')

lw = 5
e1 = diagram.line(in1, vqed, color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))
p1 = diagram.line(in2, v1pp, color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))
e2 = diagram.line(vqed, out1, color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))
pomreg = diagram.line(v1, v4,**pom_style)
p2 = diagram.line(v1p, out2, color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))
wz1 = diagram.line(vqed, vx, style='wiggly', color='green', lw=lw, nwiggles=9.5, amplitude=0.01)
O = diagram.operator([v4,v3])
#O.text("$\Sigma$")
Op = diagram.operator([v1pp,v1p])
#Op.style.update(facecolor='k')
x1 = diagram.line(v5, outx1, color='black', lw=lw,arrow=False)
x2 = diagram.line(v5, outx2, color='black', lw=lw,arrow=False)
x3 = diagram.line(v5, outx3, color='black', lw=lw,arrow=False)
X = diagram.line(vx, outX, color='black', lw=lw,arrow=False)
beta = diagram.line(vx, v3, color='black', lw=lw,arrow=False)
t    = diagram.line(v1t,v2t,**t_style)
###################################################################################
e1.text(r"$e(k)$" ,fontsize=40)
e2.text(r"$e(k')$",fontsize=40)
#p1.text(r"$p(P)$" ,fontsize=40)
#p2.text(r"$p(P')$",fontsize=40)
#pomreg.text(r"$x_{I\!P}$",fontsize=40)
#beta.text(r"$\beta$",fontsize=40)
###################################################################################
#wz1.text(r"$\gamma^*q$",fontsize=40,color='g')
diagram.text(0.42, 0.74,r"$\gamma^*\!(q)$",fontsize=40,color='g')
diagram.text(0.33, 0.43, r"$I\!R$,$I\!P$", fontsize=40)
diagram.text(0.65, 0.25, r"$p(P')$", fontsize=40)
diagram.text(0.2, 0.3, r"$p(P)$", fontsize=40)
diagram.text(0.43, 0.63, r"$\beta$", fontsize=40)
diagram.text(0.82, 0.6, r"}", fontsize=200)
diagram.text(0.93, 0.6, r"$X(M_X)$", fontsize=40)
diagram.text(0.45, 0.43, r"$x_{I\!P}$", fontsize=40)
diagram.text(0.4, 0.2, r"$t$", fontsize=40)
###################################################################################
diagram.plot()
matplotlib.pyplot.savefig('DiffDIS.pdf')
matplotlib.pyplot.show()
```
